```python
import pandas as pd 
import numpy as np 


data = pd.read_csv("./data/player_stats_202122.csv", sep=';', encoding='ISO-8859-1')
data.head(5)
```

```python
data.columns[data.isnull().any()].tolist()
```

```python
football_player_stats = pd.DataFrame()

football_player_stats['Player'] = data['Player']
football_player_stats['Nation'] = data['Nation']
football_player_stats['Squad'] = data['Squad']
football_player_stats['League'] = data['Comp']
football_player_stats['Age'] = data['Age']
football_player_stats['MP'] = data['MP']
football_player_stats['Position'] = data['Pos']
football_player_stats['Goals/90'] = data['Goals']
football_player_stats['Goals/Shot'] = data['G/Sh']
football_player_stats['Shots'] = data['Shots']
football_player_stats['TouchesDefPen'] = data['TouDefPen']
football_player_stats['TouchesDef3rd'] = data['TouDef3rd']
football_player_stats['TouchesMid3rd'] = data['TouMid3rd']
football_player_stats['TouchesAtt3rd'] = data['TouAtt3rd']
football_player_stats['TouchesAttPen'] = data['TouAttPen']
football_player_stats['BlocksOnTargetShots'] = data['BlkShSv']
football_player_stats['BlocksShots'] = data['BlkSh']
football_player_stats['BlocksPasses'] = data['BlkPass']
football_player_stats['Blocks'] = (data['Blocks'] - (data['BlkShSv'] + data['BlkSh'] + data['BlkPass'])).round(0).astype(int)
football_player_stats['PKGoals'] = ((data['ShoPK'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PKAttempted'] = ((data['PKatt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Goals'] = ((data['Goals'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Pass'] = ((data['PasTotAtt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PassCompleted'] = ((data['PasTotCmp'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PassComp%'] = ((football_player_stats['PassCompleted'] / football_player_stats['Pass'])).round(2)
football_player_stats['Assist'] = ((data['Assists'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Cross'] = ((data['PasCrs'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['CrossCompleted'] = ((data['CrsPA'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['CrossComp%'] = ((football_player_stats['CrossCompleted'] / football_player_stats['Cross'])).round(2)
football_player_stats['Tackle_Won'] = ((data['TklWon'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['SucDribble'] = ((data['DriSucc'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Dribble'] = ((data['DriAtt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['DribbleComp%'] = ((football_player_stats['SucDribble'] / football_player_stats['Dribble'])).round(2)
football_player_stats['YCards'] = ((data['CrdY'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['RCards'] = ((data['CrdR'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['FoulsCommitted'] = ((data['Fls'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['FoulsDrawn'] = ((data['Fld'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['OGoals'] = ((data['OG'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerWon'] = ((data['AerWon'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerLost'] = ((data['AerLost'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerWon%'] = ((data['AerWon'] / (data['AerWon'] + data['AerLost'])) * 100).round(2)

football_player_stats.head()
```

```python
Attack_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "Goals", "Goals/Shot", "Shots", "PKGoals", "PKAttempted","TouchesAtt3rd", "TouchesAttPen"]]

Att_weights = [0.3,0.1,0.05,0.2,0.1,0.1,0.15]
Attack_Stats["Combined"] = Attack_Stats.iloc[:,6:].mul(Att_weights).sum(axis=1)
Attack_Stats = Attack_Stats.sort_values(by='Combined', ascending=False)
Attack_Stats.head()
```

```python
# Pass_Stats = football_player_stats[["Player","Pass", "PassCompleted", "PassComp%"]]
Pass_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "PassComp%"]]
Pass_weights = [1]
Pass_Stats["Combined"] = Pass_Stats.iloc[:,6:].mul(Pass_weights).sum(axis=1)
Pass_Stats = Pass_Stats.sort_values(by='Combined', ascending=False)
Pass_Stats.head()
```

```python
Crea_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "Assist", "CrossComp%","FoulsDrawn"]]

Crea_weights = [0.5,0.3,0.2]
Crea_Stats["Combined"] = Crea_Stats.iloc[:,6:].mul(Crea_weights).sum(axis=1)
Crea_Stats = Crea_Stats.sort_values(by='Combined', ascending=False)
Crea_Stats.head()
```

```python
Tack_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "Tackle_Won", "TouchesDefPen", "TouchesDef3rd","AerWon%"]]

Tack_weights = [0.4,0.2,0.15,0.25]
Tack_Stats["Combined"] = Tack_Stats.iloc[:,6:].mul(Tack_weights).sum(axis=1)
Tack_Stats = Tack_Stats.sort_values(by='Combined', ascending=False)
Tack_Stats.head()
```

```python
Block_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "BlocksShots", "BlocksOnTargetShots","BlocksPasses"]]

Block_weights = [0.2,0.5,0.3]
Block_Stats["Combined"] = Block_Stats.iloc[:,6:].mul(Block_weights).sum(axis=1)
Block_Stats = Block_Stats.sort_values(by='Combined', ascending=False)
Block_Stats.head()
```

```python
Drib_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "DribbleComp%", "TouchesMid3rd"]]

Drib_weights = [0.8,0.2]
Drib_Stats["Combined"] = Drib_Stats.iloc[:,6:].mul(Drib_weights).sum(axis=1)
Drib_Stats = Drib_Stats.sort_values(by='Combined', ascending=False)
Drib_Stats.head()
```

```python
Indis_Stats = football_player_stats[["Player", "Nation", "Squad", "League", "Age", "Position", "YCards", "RCards", "FoulsCommitted"]]

Indis_weights = [0.2, 0.3, 0.5]
Indis_Stats["Combined"] = Indis_Stats.iloc[:,6:].mul(Indis_weights).sum(axis=1)
Indis_Stats = Indis_Stats.sort_values(by='Combined', ascending=False)
Indis_Stats.head()
```

```python
Composite_Table = pd.DataFrame()

Composite_Table["Player"] = Attack_Stats["Player"]
Composite_Table["Attack"] = Attack_Stats["Combined"]
Composite_Table["Pass"] = Pass_Stats["Combined"]
Composite_Table["Creativity"] = Crea_Stats["Combined"]
Composite_Table["Tackles"] = Tack_Stats["Combined"]
Composite_Table["Blocks"] = Block_Stats["Combined"]
Composite_Table["Dribbling"] = Drib_Stats["Combined"]
Composite_Table["Indiscipline"] = Indis_Stats["Combined"]
Composite_Table["Nation"] = football_player_stats["Nation"]
Composite_Table["Squad"] = football_player_stats["Squad"]
Composite_Table["League"] = football_player_stats["League"]
Composite_Table["Age"] = football_player_stats["Age"]
Composite_Table["Position"] = football_player_stats["Position"]
Composite_Table.head()
```
