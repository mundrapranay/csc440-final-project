```python
import pandas as pd 
import numpy as np 


data = pd.read_csv("./data/player_stats_202122.csv", sep=';', encoding='ISO-8859-1')
data.head(5)
```

```python
data.columns[data.isnull().any()].tolist()
```

```python
football_player_stats = pd.DataFrame()

football_player_stats['Player'] = data['Player']
football_player_stats['Nation'] = data['Nation']
football_player_stats['Squad'] = data['Squad']
football_player_stats['League'] = data['Comp']
football_player_stats['Age'] = data['Age']
football_player_stats['MP'] = data['MP']
football_player_stats['Position'] = data['Pos']
football_player_stats['Goals/90'] = data['Goals']
football_player_stats['Goals/Shot'] = data['G/Sh']
football_player_stats['Shots'] = data['Shots']
football_player_stats['TouchesDefPen'] = data['TouDefPen']
football_player_stats['TouchesDef3rd'] = data['TouDef3rd']
football_player_stats['TouchesMid3rd'] = data['TouMid3rd']
football_player_stats['TouchesAtt3rd'] = data['TouAtt3rd']
football_player_stats['TouchesAttPen'] = data['TouAttPen']
football_player_stats['BlocksOnTargetShots'] = data['BlkShSv']
football_player_stats['BlocksShots'] = data['BlkSh']
football_player_stats['BlocksPasses'] = data['BlkPass']
football_player_stats['Blocks'] = (data['Blocks'] - (data['BlkShSv'] + data['BlkSh'] + data['BlkPass'])).round(0).astype(int)
football_player_stats['PKGoals'] = ((data['ShoPK'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PKAttempted'] = ((data['PKatt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Goals'] = ((data['Goals'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Pass'] = ((data['PasTotAtt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PassCompleted'] = ((data['PasTotCmp'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['PassComp%'] = ((football_player_stats['PassCompleted'] / football_player_stats['Pass']) * 100).round(2)
football_player_stats['Assist'] = ((data['Assists'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Cross'] = ((data['PasCrs'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['CrossCompleted'] = ((data['CrsPA'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['CrossComp%'] = ((football_player_stats['CrossCompleted'] / football_player_stats['Cross']) * 100).round(2)
football_player_stats['Tackle_Won'] = ((data['TklWon'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['SucDribble'] = ((data['DriSucc'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['Dribble'] = ((data['DriAtt'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['DribbleComp%'] = ((football_player_stats['SucDribble'] / football_player_stats['Dribble']) * 100).round(2)
football_player_stats['YCards'] = ((data['CrdY'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['RCards'] = ((data['CrdR'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['FoulsCommitted'] = ((data['Fls'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['FoulsDrawn'] = ((data['Fld'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['OGoals'] = ((data['OG'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerWon'] = ((data['AerWon'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerLost'] = ((data['AerLost'] * data['Min']) / 90).round(0).astype(int)
football_player_stats['AerWon%'] = ((data['AerWon'] / (data['AerWon'] + data['AerLost'])) * 100).round(2)

football_player_stats.head()
```
